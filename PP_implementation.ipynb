Implementation of Supprt Vector Machine in Online Convex Programming (Online Learning)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
```

```python
class OnlineSVM:
    
    def __init__(self):
        x=1

        self.regret = 0
        self.w = 0
        self.y_predicted_value = 0

        
    def learn(self, dataset, learning_rate):
        #online SVM algorithm
        
        # initializing variables
        self.hinge_loss = 0
        self.learning_rate = learning_rate
        self.learning_rate_factor = 1/np.sqrt(self.learning_rate)
        regret_inst = np.zeros(len(dataset))
        self.y_predicted = np.zeros(len(dataset))
        self.t=1 # has to start at 1 to avoid division by zero
        self.w_star = self._calculate_exact_w(dataset)
        
        # get the first sample from the dataset
        next_sample = self._get_next_sample(dataset)
        
        # initialize the parameter vector w
        self._initialize_w(len(next_sample[0]))
        
        while next_sample:
            x_t, y_t = next_sample
            
            # calculate w_t^T * x_t
            self.y_predicted_value = np.dot(self.w,x_t)
            
            # get the sign as the predicted class for the current sample
            self.y_predicted[self.t-1] = np.sign(self.y_predicted_value)
            
            # update the sum of the hinge loss
            self.hinge_loss += max(0, 1 - y_t * self.y_predicted_value)
            
            # calculate instantaneuous regret and save in array
            regret_inst[self.t-1] = self._calculate_instantaneous_regret(next_sample)
            
            # update the parameter w
            self._update_w(x_t, y_t)
            
            # update counter
            self.t += 1
            
            # get the next sample if available
            next_sample = self._get_next_sample(dataset)
        
        
        regret = np.sum(regret_inst)
        return self.w, regret_inst, regret, self.y_predicted
    

##############################################################################################################################
# private functions

    def _get_next_sample(self, dataset):
        
        mytuple = ()
        if (self.t-1) < len(dataset):
            mytuple = (dataset[self.t-1][0:-1],dataset[self.t-1][-1])
        
        return mytuple

    
    def _initialize_w(self,size):
        # choose a random vector of given size
        self.w = np.random.rand(size)
        
        # normalize it such that the norm is less or equal than  1 / sqrt(lambda)
        if np.linalg.norm(self.w) > 1: #self.learning_rate_factor
            self.w = self.w / np.linalg.norm(self.w) # * self.learning_rate_factor
    
    
    def _update_w(self, x_t, y_t):
        
        # if this condition doesn't hold, the gradient is zero and no update of w is needed
        if (y_t * self.y_predicted_value) < 1:
            
            # update w with subgradient of the loss -y_t*x_t, the learning rate 1/lambda and annealing factor 1/t
            self.w = self.w - 1/(self.t*self.learning_rate) * (-1) * y_t * x_t

            # map w to feasibility region S
            if np.linalg.norm(self.w) > 1: #self.learning_rate_factor
                self.w = self.w / np.linalg.norm(self.w) # * self.learning_rate_factor

    def _calculate_instantaneous_regret(self, sample):
        x_t, y_t = sample
        loss_star_inst = max(0, 1 - y_t * np.dot(self.w_star, x_t))
        loss_inst = max(0, 1 - y_t * np.dot(self.w, x_t))
        
        return loss_inst - loss_star_inst
                
#    def _calculate_regret(self, dataset):
#        #w_star = self._calculate_exact_w(dataset)
#        self.loss_star = 0
#        for i in range(len(dataset)):
#            self.loss_star += max(0, 1 - dataset[i][-1] * np.dot(self.w_star, dataset[i][0:-1]))       
#        return self.hinge_loss - self.loss_star

        
    def _calculate_exact_w(self, data):
        # set initial w0
        w0 = np.ones(len(data[0][0:-1])) 
        # project it to the feasible region
        w0 = w0 / np.linalg.norm(w0) # * self.learning_rate_factor
        # define the constraint
        con = {'type': 'ineq', 'fun': lambda w : 1 - np.linalg.norm(w)}
        
        # optimize target function using scipy.optimize.minimize
        res = minimize(target_func, x0=w0, args=(data), method='SLSQP', constraints=con)
        
        return res.x


#########################################################################################################################
# not part of the class

    def target_func(w, data):
        f = 0
        for i in range(len(data)):
            f += max(0, 1 - data[i][-1] * np.dot(w, data[i][0:-1]))

        return f
```

```python
#read data
data = np.genfromtxt('toydata.csv', delimiter=',',dtype='float',skip_header=1)

#add one to each feature vector
data = np.insert(data, obj = 4, values = np.ones(len(data)), axis = 1)

# shuffle along the first axis of array
np.random.shuffle(data)
```

```python
lambda_factors = [0.05, 0.1, 1, 10, 20, 50]
accuracies = np.zeros(len(lambda_factors))
regrets = np.zeros(len(lambda_factors))
regret_insts = np.zeros((len(lambda_factors), len(data)))

for i, lambda_factor in enumerate(lambda_factors):

    myOnlineSVM = OnlineSVM()

    w_final,  regret_insts[i,:], regrets[i], y_predicted = myOnlineSVM.learn(data, lambda_factor)

    accuracies[i] = sum(y_predicted == data[:,-1]) / len(data)

```
